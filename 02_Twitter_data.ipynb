```python
import json
import bz2

data = bz2.open('./data/test.20140209-140126.json.bz2', 'rt').readlines()
```

```python
json.loads(data[101])
```

```python
tw_data = open('./data/twitter.csv', 'w')

for tw in data:
    dt = json.loads(tw)
    hs = dt['entities']['hashtags']
    hashtags = []
    for item in hs:
        hashtags.append(item['text'])
        
    tw_data.write('%s, %s, %s\n' % (dt['id'], dt['created_at'], dt['user']['time_zone']))
    tw_data.flush()
tw_data.close()
```
